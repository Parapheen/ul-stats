```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
url = 'https://ru.wikipedia.org/wiki/Категория:Списки_вузов_России'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text)
```

```python
categories_list = soup.find_all('div', {'class': 'mw-category-group'})
```

```python
categories_list
```

```python
a_s = [tag.find_all('a') for tag in categories_list]
```

```python
links = []
for i in a_s:
    for a in i:
        links.append(a.get('href'))
```

```python
len(links)
```

```python
base = 'https://ru.wikipedia.org'
```

```python
def parse_region_page(link):
    url = base+link
    page = requests.get(url).text
    soup = BeautifulSoup(page)
    
    table = soup.find('div', {'class': 'mw-parser-output'})
    trs = table.find_all('tr')
    unis = []
    for tr in trs:
        try:
            print(tr.find('td').text)
            if tr.find('td'):
                unis.append(tr.find('td').text.strip())
        except Exception as e:
            print(e)
    return unis
```

```python
unis = parse_region_page(links[1])
```

```python
res = []
for link in links[1:]:
    res.append(parse_region_page(link))
```
