```python
import requests
from bs4 import BeautifulSoup
import json
```

```python
def has_unis(page):
    print('checking')
    soup = BeautifulSoup(page)
    unis = soup.find_all('div', {'class': 'col-md-12 itemVuzPremium'})
    return unis is not None
```

```python
def parse_uni_card(page):
    print('parsing')
    soup = BeautifulSoup(page)
    unis_names = soup.find_all('div', {'class': 'itemVuzTitle'})
    return unis_names
```

```python
url = "https://vuzopedia.ru/vuz/?page"

def req(url, page):
    print('requesting')
    r = requests.get(f'{url}={page}').text
    if r:
        return r
```

```python
def main():
    res = []
    page_num = 1
    print(page_num)
    while True:
        page = req(url, page_num)
        if has_unis(page):
            unis_divs = parse_uni_card(page)
            for uni in unis_divs:
                res.append(uni.text)
            page_num += 1
            print('Res len:', len(res))
            if len(res) >= 739:
                print('done')
                break
    return res
```

```python
res = main()
```

```python
res_ = [v.strip() for v in res]
```

```python
len(res_)
```

```python
res_prod = []

for i in res_:
    res_prod.append({'name': i})
```

```python
with open('unis_rus.json', 'w') as f:
    json.dump(res_prod, f, indent=2, ensure_ascii=False)
```

```python
def encode_prod(file, name):
    res = []
    
    for line in file:
        res.append({'name': line.strip()})
    
    with open(name, 'w') as f:
        json.dump(res, f, indent=2, ensure_ascii=False)
```

```python
with open('unis_by.txt', 'r') as s:
    encode_prod(s, name ='unis_by.json')
```

```python
url_ukr = "https://ru.osvita.ua/vnz/guide/search-17-0-0-0"

def parse_ukraine(page):
    soup = BeautifulSoup(page)
    unis_table = soup.find('table', {'class': 'guide_search_result'})
    unis_names = unis_table.find_all('span', {'class': 'bhead'})
    return unis_names

def ukraine():
    res = []
    page_offset = 0
    print(page_offset)
    while True:
        r = requests.get(f'{url_ukr}-{page_offset}.html')
        unis_spans = parse_ukraine(r.text)
        for uni in unis_spans:
            res.append(uni.text)
        page_offset += 25
        print(len(res))
        if len(res) >= 484:
            print('Done with:', len(res))
            break
    return res
```

```python
res = ukraine()
```
